```python
data.Speed.plot(kind = 'line', color = 'g',label = 'Speed',linewidth='1',alpha = 0.5,grid = True)
data.Defense.plot(kind='line',color = 'r',label = 'Defense',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')


plt.xlabel('x axis')              # label = name of label
plt.ylabel('y axis')
plt.title('Line Plot')            # title = title of plot
plt.show()
```

```python
data.plot(kind='scatter', x='Attack', y='Defense',alpha = 0.5,color = 'red')

plt.xlabel('Attack')             
plt.ylabel('Defence')
plt.title('Attack Defense Scatter Plot')  
```

```python
data.Speed.plot(kind='hist',figsize=(12,12),bins=50)
plt.show()
```
